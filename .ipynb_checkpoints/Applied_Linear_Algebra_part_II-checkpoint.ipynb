## Nonsingular Matrix
### Theorem: a matrix $A \in \mathbb{R}^{m×n}$ with $m \ge n$ has full rank if and only if it maps no two distinct vectors to the same vector, and $Ax = Ay \Rightarrow x = y$.

### Singular matrix : a square matrix $A \in \mathbb{R}^{n×n}$ that is not full rank.

### Nonsingular matrix : a square matrix $A \in \mathbb{R}^{n×n}$ of full rank.

### If $A$ is nonsingular, we can uniquely express any vector $y \in \mathbb{R}^n$ as $y=Ax$ for some unique $x \in \mathbb{R}^n$. If we chose vectors $e_i =Ax_i, i = 1,...,n$, we can write
### $AX=A[x_1 ··· x_n]=[Ax_1 ··· Ax_n]=[e_1 ··· e_n]=I$

```python
matrix1 = np.matrix([[1,2,3],[7,2,5],[9,1,6]])
print("Matrix (full rank would be 3):")
print(matrix1)
print("\nRank:")
print(np.linalg.matrix_rank(matrix1))

print("\nTherefore, nonsingular")

matrix2 = np.matrix([[1,2,3],[7,2,5],[2,4,6]])
print("\nMatrix (full rank would be 3):")
print(matrix2)
print("\nRank:")
print(np.linalg.matrix_rank(matrix2))

print("\nTherefore, singular")
```

## Matrix Inverse
### Matrix inverse: the inverse of a square matrix $A \in \mathbb{R}^{n×n}$ is the matrix $B \in \mathbb{R}^{n×n}$ such that
### $AB = BA = I$ where $I$ is the identity matrix
### The inverse of $A$ is denoted by $A^{−1}$.
### Any square, nonsingular matrix $A$ has a unique inverse $A^{−1}$ satisfying $AA^{−1} = A^{−1}A = I$.


```python
print("Matrix:")
print(matrix1)

print("\nMatrix inverse:")
print(np.linalg.inv(matrix1))

print("\nMatrix * Matrix_inverse")
print(matrix1 * np.linalg.inv(matrix1))
```

### $(AB)^{-1} = B^{−1}A^{−1}$ (assuming both inverses exist).


```python
print("Matrix A:")
print(matrix1)

matrix2 = np.matrix([[8,3,6],[9,1,8],[6,5,2]])
print("\nMatrix B:")
print(matrix2)

print("\nTest for equality")
print(np.linalg.inv(matrix1*matrix2))
print(np.linalg.inv(matrix2)*np.linalg.inv(matrix1))
```

### $(A^T )^{−1} = (A^{−1} )^T$

```python
print("Matrix A:")
print(matrix1)

print("\nTest for equality")
print(np.linalg.inv(matrix1.transpose()))
print(np.linalg.inv(matrix1).transpose())
```

### If $Ax = b$, then $A^{−1}b$ gives the vector of coefficients in the linear combination of the columns of $A$ that yields $b$:

### $b=Ax=x_1a_1 +···+x_na_n, x=A^{-1}b= \begin{bmatrix} x_1 \\ \dots \\ x_n \end{bmatrix}$

```python
print("Matrix A:")
print(matrix1)

print("\nVector of coefficients if b=[0,0,0]")
print(np.linalg.inv(matrix1)*np.matrix([0,0,0]).transpose())

print("\nVector of coefficients if b=[1,0,0]")
print(np.linalg.inv(matrix1)*np.matrix([1,0,0]).transpose())
```

## Invertible Matrix Theorem
### Let $A \in \mathbb{R}^{n×n}$ be a square matrix, then the following statements are equivalent:
* $A$ is invertible, $A$ is nonsingular.
* There exists a matrix $A^{−1}$ such that $AA^{−1} = I_n = A^{−1}A$.
* $Ax = b$ has exactly one solution for each $b \in \mathbb{R}^n$.
* $Az=0$ has only the trivial solution $z=0$,i.e. $N(A)=∅$
* The columns of $A$ are linearly independent.
* $A$ is full rank, i.e. rank($A$) = $n$.
* $det(A)\ne 0$.
* 0 is not an eigenvalue of $A$.

* * *

## One of the fundamental problems in linear algebra: 

### Given $A\in \mathbb{R}^{m×n}$ and $b \in \mathbb{R}^m$, find $x \in \mathbb{R}^n$ such that $Ax = b$.
### The set of solutions of a linear system can behave in three ways:
1. The system has one unique solution.
2. The system has infinitely many solutions.
3. The system has no solution.

***

## For a solution $x$ to exist, we must have $b \in R(A)$.

#### Where R( ) is the row space, m is the row dim, and n is the column dim

* If $b \in R(A)$ and rank($A$) = n, then there is a unique solution (usually square systems: $m = n$).
* If $b \in R(A)$ and rank($A$) $\lt n (N (A) \ne ∅)$, then there are infinitely many solutions (usually underdetermined systems: $m < n$).
* If $b \notin R(A)$, then there is no solution (usually overdetermined systems: $m > n$).

# Gaussian Elimination (row reduction)

How do we solve $Ax = b$, $A \in \mathbb{R}^{n×n}$?
For some types of matrices, $Ax = b$ is easy to solve.
* Diagonal matrices: equations independent, so $x_i = b_i/a_{ii}$, $\forall i$
* Upper/Lower triangular matrices: back/forward substitution.

Gaussian Elimination: Transform the system $Ax = b$ into an upper triangular system $Ux = y$ using elementary row operations.
The following are elementary row operations that can be applied to the augmented system $[A | b]$ to introduce zeros below the diagonal without changing the solution set x:

1. Multiply row by a scalar.
2. Add scalar multiples of one row to another.
3. Permute rows.

***

## Gaussian Elimination and LU Decomposition

### We can think of the elementary row operations applied to A as linear transformations, 
### which can be represented as a series of lower triangular matrices, $L_1, · · · , L_{n−1}$.
### $L_{n−1} ···L_1A = U$ or $L^{−1}A = U, L^{−1} = L^{−1}_1 ··· L^{−1}_{n−1}$

### rearranging yields $A = LU$.

### If we can factor $A$ as $A = LU$, then we can solve the system
### $Ax = b$ by solving two triangular systems:

* Solve $Ly = b$ using forward substitution,
* Solve $Ux = y$ using backward substitution.

### Gaussian elimination implicitly computes the LU factorization.

```python
print("Matrix A:")
print(matrix1)

print("\nLU Decomposition into L and U, where L is permuted")
L, U = lu(matrix1, permute_l=True)

print("\nL")
print(L)

print("\nU")
print(U)

print("\nL dot product U")
print(L.dot(U))

print("\nLU Decomposition into P, L and U, where L is not permuted")
P, L, U = lu(matrix1, permute_l=False)
print("\nP")
print(P)

print("\nL")
print(L)

print("\nU")
print(U)

print("\nP.dot(L.dot(U))")
print(P.dot(L.dot(U)))
```

## Partial Pivoting

### Pure Gaussian elimination can not be used to solve general linear systems because we may encounter a zero on the diagonal during the process, e.g.

### $ A = \begin{bmatrix} 0 & 1 \\
                          1 & 2 \end{bmatrix}$

### In practical implementations of the algorithm, the rows of the matrix must be permuted to avoid division by zero (see examples with and without permutation above).

### Partial pivoting swaps rows if an entry below the diagonal of the current column is larger in absolute value than the current diagonal entry (pivot element).

***
## Permutations

### The reordering, or permutation, of the columns in partial pivoting can be represented as a linear transformation.
### Permutation matrix : a square, binary matrix $P \in \mathbb{R}^{n×n}$ that has exactly one entry 1 in each row and each column.

### Left-multiplication of a matrix $A$ by a permuation matrix reorders the rows of $A$, while right-multiplication reorders the columns of $A$.

For example, let $P = \begin{bmatrix} 0 & 1 & 0 \\
                                      0 & 0 & 1 \\
                                      1 & 0 & 0 \end{bmatrix}$, and $A = \begin{bmatrix} 1 & 2 & 3 \\
                                                                                     10 & 20 & 30 \\
                                                                                     100 & 200 & 300 \end{bmatrix}$,

### then the row and column permutations are, respectively:

$PA = \begin{bmatrix} 10 & 20 & 30 \\
                      100 & 200 & 300 \\
                      1 & 2 & 3 \end{bmatrix}$, and $AP = \begin{bmatrix} 3 & 2 & 1 \\
                                                                          30 & 20 & 10 \\
                                                                          300 & 200 & 100 \end{bmatrix}$

## LU Decomposition

### Let $A \in \mathbb{R}^{n×n}$ be a square matrix, then the LU decomposition factors $A$ into the product of a lower triangular matrix $L \in \mathbb{R}^{n×n}$ and an upper triangular matrix $U \in \mathbb{R}^{n×n} A = LU$.

### When partial pivoting is used to permute the rows, we obtain a decomposition for the form $L_{n−1}P_{n−1} · · · L_1P_1A = U$, which can be written in form: $L^{−1}PA = U$.

### In general, any square matrix $A \in \mathbb{R}^{n×n}$ (singular or nonsingular) has a factorization
### $P A = LU$, where $P$ is a permutation matrix.

## Cholesky and Positive Definiteness

### Positive definite matrix: a matrix $A \in \mathbb{R}^{n×n}$ is positive definite if $z^TAz \gt 0$ for every $z \in \mathbb{R}^n$, often denoted by $A\succ 0$.

### To determine if a matrix is positive definite:

### $A = \begin{bmatrix} 4 & -2 \\ -2 & 10 \end{bmatrix}$

### $x^T Ax = \begin{bmatrix} x_1 & x_2 \end{bmatrix} \begin{bmatrix} 4 & -2 \\ -2 & 10 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}$

### $= 4x_1^2 - 4x_1x_2 + 10x_2^2$

### $= (2x_1 - x_2)^2 +9x_2^2 \gt 0$, $\forall x \ne 0$

### That is to say, any values selected for $x_1$ and $x_2$ must result in a positive value for $x^T Ax$, and all $x$s cannot be zero.

* Positive semi-definite matrix: $z^T Az \ge 0$ for every $z \in \mathbb{R}^n$.
* Negative definite matrix: $z^T Az \lt 0$ for every $z \in \mathbb{R}^n$.

### A symmetric matrix $A$ is positive definite ($A \succ 0$) if and only if there exists a unique lower triangular matrix $L$ such that $A = LL^T$ .

### This factorization is called the Cholesky decomposition of $A$.

### The Cholesky algorithm, a modified version of Gauss elimination, is used to compute the factor $L$. It can be approximately twice as efficient as LU decomposition.

### Basic idea: In Gauss-elimination we use row operations to introduce zeros below the diagonal in each column. To maintain symmetry, we can apply the same operations to the columns of the matrix to introduce zeros in the first row.

```python
print("Matrix must be symmetric and positive definite")
matrix3 = np.matrix([[4,12,-16],[12,37,-43],[-16,-43,98]])

print("Matrix A:")
print(matrix3)

print("\nL")
L = np.linalg.cholesky(matrix3)
print(L)

print("\nLL-transposed")
print(L.dot(L.transpose()))
```

## Orthogonality

### Recall that two vectors $ x, y \in \mathbb{R}^n $ are orthogonal if $ x^T y = 0 $ (their dot product is zero), and a set of vectors {$ u_1, \dots , u_n $} is orthonormal if

### $ u^T_iu_j= \begin{cases}
    1, & \text{for } i=j\\
    0, & \text{for } i\ne j
\end{cases} $, $ \|u_i\|=1,\forall i $

### Orthogonal (Unitary) Matrix : a matrix $ Q \in \mathbb{R}^{n×n} $ is orthogonal if its columns $ q_1, \dots , q_n \in \mathbb{R}^n $ form an orthonormal set:

## $ QQ^T = \begin{bmatrix} q_1^T \\ \dots \\ q_n^T \end{bmatrix} \begin{bmatrix} q_1 & \dots & q_n \end{bmatrix} = I_n $

### If $ Q $ is orthogonal, then $ Q^{−1} = Q^T $ (inverse is easy to compute!) 

### Orthogonal transformations preserve lengths and angles and represent rotations or reflections.

***
## Projectors

### Recall our notation for the projection of vector $v$ onto $u$: 

### $proj_uv=\frac{u^Tv}{\|u\|} \frac{u}{\|u\|}$

### We can generalize the idea to projections onto subspaces of $\mathbb{R}^n$.
### Projector: a square matrix $P \in \mathbb{R}^{n×n}$ is a projector if $P = P^2$.

* For $v \in \mathbb{R}^n$, $Pv$ represents the projection of $v$ into the range of $P$.
* If $v \in R(P)(\exists x:v=Px$),then $Pv=P^2x=Px=v$.

***
## Orthogonal Projectors

### If $x \in \mathbb{R}^m$ and $A\in \mathbb{R}^{m×n}$ is a matrix whose columns form a subspace of $\mathbb{R}^m$, then $x$ can be decomposed as:

### $ x=x^{\|}+x^\perp$ , where $x^{\|} \in R(A)$ and $x^\perp \in R(A)^\perp$ 

### $x^{\|}$ is the orthogonal projection of $x$ onto $R(A)$.

### $x^{\|}$ is the vector in $R(A)$ “closest” to $x$, in the sense that $\|x−x^{\|}\| \lt \|x−v\|, \forall v \in R(A) \ne x^{\|}$.

### An orthogonal projector is a projector for which the range and nullspace are orthogonal complements.

### An orthogonal projector satisfies $P^T = P$.

### If $u \in \mathbb{R}^n$ is a unit vector, then $P_u = uu^T$ is the orthogonal projection onto the line containing $u$.

### If $u_1, \dots , u_k$ are an orthonormal basis for subspace $V$ and $A=[u_1 \dots u_n]$,then the projection onto $V$ is 

### $P_A = AA^T$

### or equivalently: 
### $proj_Ax = P_Ax $
### $= (u^T_1 x)u_1 + \dots + (u^T_n x)u_n$ 
### $= proj_{u_1}x + \dots + proj_{u_n}x$

## Least Squares

### We want to solve a system of linear equations, $Ax = b$, but what do we do if $b \notin R(A)$ (i.e. there does not exist a vector x such that $Ax = b$)?

### For example, let $A \in \mathbb{R}^{m \times n}$, $m \gt n$ be a tall-and-skinny (overdetermined) matrix. Then for most $b \in \mathbb{R}^m$, there is no solution $x \in \mathbb{R}^n$ such that $Ax = b$.

### Least Squares problem: Define residual $r = b - Ax$ and find the vector $x$ that minimizes
$\|r\|_2^2 = \|b - Ax\|_2^2$.

### We can decompose any vector $b \in \mathbb{R}^m$ into components $b = b_1 + b_2$, with $b_1 \in R(A)$ and $b_2 \in N(A^T)$.
### Since $b_2$ is in the orthogonal complement of the $R(A)$, we obtain the following expression for the residual norm:
### $\|r\|_2^2 =\|b_1 - Ax + b_ 2\|_2^2 =\|b_1 - Ax\|_2^2 +\|b\|_2^2$ 

### which is minimized when $Ax = b_1$ and $r = b_2 \in N(A^T )$.

### The least squares solution $x$ occurs when $r = b_2 \in N (AT )$ or equivalently $A^T r = A^T (b - Ax) = 0$.

### Rearranging this expression gives the normal equations: $A^TAx = A^Tb$

### If $A$ has full column rank, then $\tilde{A} = A^T A$ is invertible. 

### Thus $\tilde{A}x = \tilde{b}$, where $\tilde{b} = A^T b$ has a unique solution $x$.

### If $x^* \in \mathbb{R}^n$ is the least squares solution to $Ax = b$, then $Ax^*$ is the orthogonal projection of b onto $R(A)$.

### Combining the solution $x^* = (A^T A)^{-1}A^T b$ and $Ax^* = P_Ab$ gives the matrix for the orthogonal projection onto $R(A)$:

### $P_A = A(A^T A)^{-1}A^T$

## Ways to Calculate Least Squares

1. QR Decomposition

The normal equation AT Ax = AT b can be rewritten using the QR factorization A = QR:
ATAx = ATb RTQTQRx = RTQTb RTRx = RTQTb
Rx = QTb
Solution using QR decomposition:
I compute QR factorization of A : A = QR I formy=QTb
I solve Rx = y by back substitution

2. Cholesky factorization

M = AT A is symmetric and positive definite, and can be rewritten in terms the Cholesky factorization M = LLT
Solution using Cholesky factorization:
I formM=ATA
I compute the Cholesky factorization of M: M = LLT I formy=ATb
I solve Lz = y by forward substitution
I solve LT x = z by back substitution
